```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
train = pd.read_csv("train.csv", index_col="index")
test = pd.read_csv("test.csv", index_col="index")
```

train = train.dropna()

```python
train = train.drop(columns=["address", "city", "image_id"])
test = test.drop(columns=["address", "city", "image_id"])
```

```python
train
```

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model = linear_model.LinearRegression()
```

```python
y = train["retail_price"]
X = train.drop(["retail_price"], axis = 1)
```

```python
model.fit(X, y)
```

```python
test
```

```python
pred = model.predict(test)
```

```python
test["retail_price"] = pred
test
```

```python
test.drop(["rooms", "sqft","toilets","predicted"], axis = 1, inplace=True)
```

```python
test.to_csv("pred.csv")
```
